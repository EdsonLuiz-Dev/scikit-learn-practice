```python
# you can also generate data with the make functions:
from sklearn.datasets import make_blobs # this creates clusters. There are many different make functions with many different shapes. Use this to test and visualize data.
```

```python
X, y = make_blobs(n_samples=5000, centers=5) # centers in this case are the number of clusters
# X, y = make_blobs(n_samples=5000, centers=5, random_state=0) - you can make this stop generating random values every time by adding the random state parameter.
# On shaped blobs, you can add the noise=n parameter. N being any value between 0 and 1.
```
